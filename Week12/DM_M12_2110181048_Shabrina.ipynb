```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#1 tampilkan transaction.csv
dataset = pd.read_csv("transaction.csv")
dataset
```

```python
#2 ambillah data pada dataset untuk fitur Qty, Country (“Germany”), month, year (“2011”)

dataset['InvoiceDate'] = pd.to_datetime(dataset['InvoiceDate'])
data = dataset[(dataset['InvoiceDate'].dt.year == 2011) & (dataset['Country'] == 'Germany')].reset_index()
data
```

```python
#3 ambillah month dari data dan akumulasi Qty pada bulan yang sama, dan tampilkan

data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
totalQty = data.groupby(data['InvoiceDate'].dt.month)['Qty'].sum().reset_index()
totalQty.columns = ['Month', 'TotalQty']
totalQty
```

```python
#4 Visualisasikan pergerakan nilai TotalQty dimana sumbu x=month dan sumbu y=total Qty

plt.figure(figsize=(12,6))
plt.plot(totalQty['Month'], totalQty['TotalQty'], color = 'black', marker='o')
```
